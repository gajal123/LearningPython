**Protocols** are informal interfaces. They cannot be enforced by formal interfaces can. We can implement a protocol by implementing the method expected by it

## Formal Interfaces: ABCs

2 classes can have the same protocol defined which may have the same meaning.
Eg: an aeroplane and bird both can fly() but they are not the same thing even if they implement the same interface/protocol. ABC's can help solve this issue.

In ABC, we define Base class which is abstract. We define methods on this class as abstarct methods.Hence any  class deriving from this has to implement those.

```
import abc

class Bird(abc.ABC):
  # define an abstract method
  @abc.abstractmethod
  def fly():
    pass

# Now any class deriving from Bird will have to define fly()

class Parrot(Bird):
  pass

p = Parrot()
```

```
# to fix the issue, we define fly on Parrot

class Parrot(Bird):
  def fly():
    print(f"inside Parrot fly")

p = Parrot()
print(isinstance(p, Bird))
```
