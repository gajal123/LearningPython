*   solving all sub-problems one by one in order to reach the best possible solution.
*   uses a recursive approach
*   Backtracking is a systematic way of trying out different sequences of decisions until we find one that "works." 
*   Backtracking works in incremental way and is an optimization over the Naive solution where all possible configurations are generated and tried.
*   expensive. usually O(n!)
