```
class newNode:
    def __init__(self, data):
        self.left = None
        self.right = None
        self.data = data
```

### Insertion:

- always insert as leaf[link text](https://) node

## Deletion

- Lead node: simply remove from tree
- 1 child: copy child to node and delete.
 - 2 children: Find inorder successor of node. replace node with inorder successor/ or predecessor and delete 

## checking BST
1. traverse every node once and keep track of min, max allowed.
2. Do in-order traversal and store result in auxiliary array. If array is sorted, BST. 
Use of auxiliary node can be avoided. check every previous visited node is smaller than current

```
def isBST(node, min, max):
    if not node:
        return True
    print(node.data, 'min: ', min, ' max: ', max)
    if min and node.data < min or max and node.data > max:
        return False
    return isBST(node.left, min, node.data) and isBST(node.right, node.data, max)

   
```
