Given an array, find all triplets such that the sum of 2 is equal to the third element

Step 1: Sort the aarray

```
def triplets_with_sum(arr):
  arr.sort()
  result = []
  i = len(arr) - 1
  j = 0
  k = i-1
  while(i > 1):
    tot = arr[j] + arr[k]
    if tot == arr[i]:
      result.append([arr[i], arr[j], arr[k]])
      i -= 1
      j = 0
      k = i-1
    elif tot < arr[i]:
      j += 1
    else:
        k -=1
    if j >= k:
      i -= 1
      j = 0
      k = i-1
  return result
        
  
print('[1, 3, 6, 6, 9, 12]: ',triplets_with_sum([1, 3, 6, 6, 9, 12]))
print('[5, 32, 1, 7, 10, 50, 19, 21, 2]: ',triplets_with_sum([5, 32, 1, 7, 10, 50, 19, 21, 2]))
print('[5, 32, 1, 7, 10, 50, 19, 21, 0]: ',triplets_with_sum([5, 32, 1, 7, 10, 50, 19, 21, 0]))
```
