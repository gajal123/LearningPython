**Assert**
 - The proper use of assertions is to inform developers about unrecoverable errors in a program
 - Act as internal self checks
 - Should not be used for Data validations
 - can be globally disabled with interpreter settings

```
# Caveats: tuples evaluate to true
# assert(1==2), 'This will fail'
assert(1==2, 'This should fail but won"t')
```
