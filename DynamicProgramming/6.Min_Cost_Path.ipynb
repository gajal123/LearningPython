###Given a Matrix, with positive costs, find the minimum cost to reach (m,n) from (0,0). 
 You can only traverse down, right and diagonally lower cells from a given cell,

Step 1: Create a m*n matrix to store costs. Initialize 1st row and 1st column with sum upto that index.

Step2: For every remaining cell, min_path will be minimum of (cell_above, cell_right, cell_diagonally_up) + cost of current cell.

###Time Complexity of the DP implementation is O(mn) which is much better than Naive Recursive implementation.

```
from copy import deepcopy
def minCost(cost, m, n):
  MC = deepcopy(cost)
  for i in range(1, m):
    MC[i][0] += MC[i-1][0]
  for j in range(1, n):
    MC[0][j] += MC[0][j-1] 
  for i in range(1, m):
    for j in range(1, n):
      MC[i][j] = min(MC[i-1][j], MC[i-1][j-1], MC[i][j-1]) + cost[i][j]
  return MC[m-1][n-1]
```

```
# Case 1
cost= [ [1, 2, 3], 
        [4, 8, 2], 
        [1, 5, 3] ] 
min_cost = minCost(cost, 3, 3)
print(f'Min Cost: {min_cost}')
```

```
# CASE 2
cost= [ [31, 100, 65, 12, 18], 
        [10, 13, 47, 157, 6], 
        [100, 113, 174, 11, 33],
        [88, 124, 41, 20, 140],
        [99, 32, 111, 41, 20],
      ] 

min_cost = minCost(cost, 5, 5)
print(f'Min Cost: {min_cost}')
```
