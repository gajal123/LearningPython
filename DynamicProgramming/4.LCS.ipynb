##Longest Common Subsequence
LCS for input Sequences **ABCDGH** and **AEDFHR** is **ADH** of length 3.

LCS for input Sequences **AGGTAB** and **GXTXAYB** is **GTAB** of length 4.

```
def lcs_recursive(st1, st2, l1, l2):
  if l1 == 0 or l2 == 0:
    return 0
  if st1[l1-1] == st2[l2-1]:
    return lcs_recursive(st1, st2, l1-1, l2-1) + 1
  return max(lcs_recursive(st1, st2, l1-1, l2),lcs_recursive(st1, st2, l1, l2-1))
  
  
st1 = 'ABCDGH'
l1 = len(st1)
st2 = 'AEDFHR'
l2 = len(st2)

print(f'LCS uding Recursion: {lcs_recursive(st1, st2, l1, l2)}')
```

```
def lcs_dp(st1, st2):
  l1 = len(st1)
  l2 = len(st2)
  LCS = [[0 for i in range(l2+1)] for j in range(l1+1)]
  for i in range(1,l1+1):
    for j in range(1,l2+1):
      if st1[i-1] == st2[j-1]:
        LCS[i][j] = LCS[i-1][j-1] + 1
      else:
        LCS[i][j] = max(LCS[i-1][j], LCS[i][j-1])
  return LCS[-1][-1]
  

st1 = 'ABCDGHWWX'
st2 = 'AEDFHR'
print(f'LCS using DP: {lcs_dp(st1, st2)}')
```
