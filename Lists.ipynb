```
lst = [1, 4, 3, 6, 10, 5, 9]
```

```
rev_lst = [i for i in reversed(lst)] # Does not change original list
print('Reversed: ', rev_lst)
print('Original list: ', lst)
print('Index of 4:', lst.index(6))
```

```
print('Sorted: ', sorted(lst)) # does not change original list
print('Original list: ', lst)

print('Sorted: ', lst.sort()) # Changes original list. 
print('.sort() returns None to remind us that it changed the target object in place and did not create a new list.')
print('II. Reverse sort {}'.format(sorted(lst, reverse=True)))
print('Original list: ', lst)
```

```
print('SORTING TUPLES')
tup = [('john', 2, 'D'),
       ('alex', 4, 'B'),
       ('john', 2, 'C')
      ]
tup.sort(key=(lambda s: s[2]))
tup
```

```
print('SORTING CLASS')
class Student:
  def __init__(self, name, grade, age):
    self.name = name
    self.grade = grade
    self.age = age
  def __repr__(self):
    return repr((self.name, self.grade, self.age))

student_objects = [
    Student('john', 'A', 15),
    Student('jane', 'B', 12),
    Student('dave', 'B', 10)
]
print('students_objects: ', student_objects)
student_objects.sort(key=lambda x:x.age)
print('students_objects sorted by age: ', student_objects)
```

Python .sort() and sorted() uses Timsort algo (by Tim Peters), that switches between insertion, merge,etc, depending on how ordered the data is. 

```
new_list = ['abcrf', 'Cdef', 'abc', 'ghrts']
print('Sorted: {}'.format(sorted(new_list, key=str.lower)))
print('Sorted by length: {}'.format(sorted(new_list, key=len)))
print('Reverse Sorted by length: {}'.format(sorted(new_list, key=len, reverse=True)))
```

```
print('BINARY search using bisect')
import bisect
lst = [1, 3, 5, 7, 13, 15, 18] # has to be sorted
print('Position for inserting 2: {}'.format(bisect.bisect(lst, 2))) # bisect is the same as bisect_after
print('Position for inserting 1: {}'.format(bisect.bisect(lst, 1)))
print('Position for inserting 1 before: {}'.format(bisect.bisect_left(lst, 1)))
```

```
print('Insertion using bisect')
bisect.insort(lst, 6)
print('Inserted after 5')
lst
```
