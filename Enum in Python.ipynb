Enum is a set of symbolic names vound to unique, constant values
 - Member value can be anything - int, str, etc
 - Enum members are hashable and hence can be used in dictionaries and sets
 - cannot create 2 members with the same name

```
from enum import Enum
class Color(Enum):
  RED = 1
  YELLOW = 3
  GREEN = 2

# Printing Enum as string
print(Color.RED.name)
```

The class Color is an Enum or Enumeration
.name is used to access value.


Enums are iterable

```
for color in Color:
  print(color)
```

Enuma can be accessed in 2 ways: name and value



```
print(Color.RED.name, Color.RED.value)
```

Comparison
- **is** or **is not**
- **==** or **!=**
- ordered comparison like **>** are not supported

```
print(f"Enum member associated with value 2 {Color(2)}")
print(f"Enum value associated with member YELLOW:  {Color['YELLOW'].name}")
print(f"Green and yellow are same colors: {Color.GREEN is Color.YELLOW}")
```

To ensure unique values, use decorator **unique**

```
from enum import Enum, unique
class Animal(Enum):
  DOG = 1
  CAT = 2
  LION = 1

@unique
class UniqueAnimal(Enum):
  DOG = 1
  CAT = 2
  LION = 1
```

Using automatic values:

If the exact value is unimportant, use auto. The values are chosen by _generate_next_value_(), which can be overridden:

```
from enum import Enum, auto
class ColorAuto(Enum):
  RED = auto()
  BLUE = auto()
  GREEN = auto()

list(ColorAuto)
```

__members__ is a read only mapping of name to members

```
for name, member in Color.__members__.items():
  print(f"Name: {name}, Member: member")
```
