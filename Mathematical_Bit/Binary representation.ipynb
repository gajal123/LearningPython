### Binary Left Shift and Binary Right Shift

**Left shift:** Multiplies the number by 2

**Right shift:** Divides the number by 2

This is the same as shifting a decimal number. On shifting left, 10 becomes 100 (due to base 10). Similarly 7 becomes 14 on left shift due to the base being 2. Essentially, every bit gets multiplied by 2.

```
num = 10
bin_num = bin(num)
print(f"Left shifting {num}: {num << 1}")
print(f"Right shifting {num}: {num >> 1}")
```
