A number is a multiple of 3 if in it's binary represenattion, (no of set odd bits) - (no of set even bits) is a multipke of 3

```

def mul_3(n):
  if n%3 == 0:
    return True
  return False

def mul_3_better(n):
  odd_count = 0
  even_count = 0

  # Make no positive if + n is multiple of 3 
  # then is -n. We are doing this to avoid 
  # stack overflow in recursion 
  if(n < 0):  
      n = -n 
  if(n == 0): 
      return 1
  if(n == 1):  
      return 0

  while(n): 

      # If odd bit is set then 
      # increment odd counter  
      if(n & 1):  
          odd_count += 1

      # If even bit is set then 
      # increment even counter  
      if(n & 2): 
          even_count += 1
      n = n >> 2

  return mul_3_better(abs(odd_count - even_count)) 
```

```
n = 100
%timeit x = mul_3(n)
print(x)
%timeit y = mul_3_better(n)
print('is_multiple: ',bool(y))
```
