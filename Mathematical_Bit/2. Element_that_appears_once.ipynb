###1. Given an array where every element occurs three times, except one element which occurs only once. Find the element that occurs once

```
from collections import Counter
import sys
arr = [1, 1, 1, 2]
c = Counter(arr)
print(c)

sys.getsizeof(c)

```

Counter can increase the space complexity. Another way would be using bit algos
Concept:
  1. get bit representation of number. 
  2. For every position, store the (sum of all numbers %3) for that position.
  3. This sum will give you the required number
  eg:
  1: 0001
  2: 0010
  index 1 sum: (1 + 1+ 1+0) %3 = 0
  index 2 sum: (0 + 0+ 0 + 1)%3 = 1
  index 3 sum: (0 + 0 + 0 + 0)%3 = 0
  index 4 sum: (0 + 0 + 0 + 0)%3 = 0
  Final number: 0010

```
INT_SIZE = 32
arr = [1, 1, 1, 20, 5, 5, 5, 10, 20, 20, 4, 4, 4]
sum = format(0, '032b')
for num in arr:
  num_binary = format(num, '032b')
  for j in range(INT_SIZE-1, -1, -1):
    if int(num_binary[j]):
      sum = sum[0:j] + str((int(sum[j]) + int(num_binary[j])) % 3) + sum[j+1:]
print('Sngle occuring number: ', int(sum, 2))

```

###2.Given an array of size N-1 and given that there are numbers from 1 to N with one element missing, the missing number is to be found.

```
from functools import reduce
def get_missing_number(arr, n):
  ans = reduce((lambda x,y:(x^y)), arr) ^ reduce((lambda x,y:x^y), range(n))
  return ans

arr = [1,2,3,5,4, 6, 7, 8, 9, 11, 12]
n = len(arr) + 2
ans = get_missing_number(arr, n)
print(f'Missing number: {ans}')
```

###3. Find the majority element in the array. A majority element in an array A of size N is an element that appears more than N/2 times in the array.

```
from collections import Counter
def get_majority_element(arr):
    n = len(arr)
    ct = Counter(arr)
    found = -1
    for key, value in ct.items():
        if value > n//2:
            found = key
            break
    return found
arr = [2,2,2,2,4,5,3,2,4,4, 2, 2]
ans = get_majority_element(arr)
print(f'Majority element {ans}')
```
