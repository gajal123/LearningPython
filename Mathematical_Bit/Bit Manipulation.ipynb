**Q1**. An array with all numbers occuring twice except one number:

**Ans**: Xor all elements of array. all duplicates would give 1. 1 ^ the required number would give the number itself

Eg: 


> [10, 10, 2, 20, 4, 2, 20]

> (10^10) ^ (20 ^ 20) ^ (2 ^ 2) ^ 4

> 0 ^ 0 ^ 0 ^ 0 ^ 4 

> 4

**Q3. ** Sum of 2 numbers without using arithmetic opeartors
**Ans: **
Step1: XOR of x and y give sum.
step 2: AND of x and y give carry. 
Step 3: left shift carry by 1 (to bring it to correct position)
Step 4: Add the shifted carry to x

```
def Add(x, y):
  while(y != 0):
    carry = (x & y)
    x = x ^ y
    y = carry << 1
    print(x, y)
  return x


print(f"Sum of 15 and 32 : {Add(15, 32)}")
print(f"Sum of 15 and 48 : {Add(15, 48)}")
print(f"Sum of 15 and 49 : {Add(15, 49)}")
```
